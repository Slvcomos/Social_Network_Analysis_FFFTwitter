```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
import networkx as nx
from operator import itemgetter
```

```python
path = "../Dataset SuperUsers/edges.csv"
```

```python
G = nx.read_edgelist(path, comments = "S", delimiter = ",", data=(("Count", int),))
degree_dict = dict(G.degree(G.nodes()))
nx.set_node_attributes(G, degree_dict, 'degree')
sorted_degree = sorted(degree_dict.items(), key = itemgetter(1), reverse = True)
```

```python
ids = []                      #tag
count = []                    #quante volte appare
named = []                    #in quale categoria finir√†
original = []                 #nome originale
for i in sorted_degree:
    ids.append(i[0])
    count.append(i[1])
    named.append("unamed")
    original.append(i[0])
```

```python
count
```

```python
def select_from_key(key, data, expection):
    al=[]
    for i in data:
        if key in i[0]:
            if len(expection)>0:
                is_except = False
                for ex in expection:
                    if ex in i[0]:
                        is_except = True
                if not is_except:
                    al.append(i[0])
            else:
                al.append(i[0])
    return al
```

```python
def replace_name(old_name, new_name, old_dataset, expection):
    for i in range(len(old_dataset)):
        if old_name in old_dataset[i]:
            if len(expection)>0:
                is_except = False
                for ex in expection:
                    if ex in old_dataset[i]:
                        is_except = True
                if not is_except:
                    if "unamed" in named[i]:
                        named[i] = new_name
                        original[i] = new_name
            else:
                if "unamed" in named[i]:
                    named[i] = new_name
                    original[i] = new_name
```
